```python
import numpy as np
import matplotlib.pyplot as plt

def low_pass_filter(signal, cutoff_freq, sampling_rate):
    # Compute the DFT of the signal
    dft = np.fft.fft(signal)
    
    # Determine the frequency values corresponding to the DFT components
    freq_values = np.fft.fftfreq(len(signal), d=1/sampling_rate)
    
    # Create a mask for frequencies below the cutoff frequency
    mask = np.abs(freq_values) <= cutoff_freq
    
    # Apply the mask to the DFT to keep only low-frequency components
    dft_filtered = dft * mask
    
    # Compute the inverse DFT to get the filtered signal
    filtered_signal = np.fft.ifft(dft_filtered).real
    
    return filtered_signal

# Example usage:
# Generate a sample noisy signal
sampling_rate = 1000  # Sampling rate in Hz
t = np.linspace(0, 1, sampling_rate, endpoint=False)
signal = 5 * np.sin(2 * np.pi * 50 * t) + 2 * np.sin(2 * np.pi * 120 * t) + 0.5 * np.random.randn(sampling_rate)

# Set the cutoff frequency for the low-pass filter (in Hz)
cutoff_frequency = 100

# Apply the low-pass filter to the signal to reduce noise
filtered_signal = low_pass_filter(signal, cutoff_frequency, sampling_rate)

# Plot the original and filtered signals
plt.figure(figsize=(12, 6))
plt.plot(t, signal, label='Original Signal')
plt.plot(t, filtered_signal, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
